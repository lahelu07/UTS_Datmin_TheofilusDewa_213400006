- Jumlah Rumah (jumlah_rumah_encod): Distribusi data ini juga cukup seragam, dengan rentang yang mencakup variasi jumlah rumah yang besar di berbagai daerah. Tidak ada nilai ekstrem, yang menunjukkan bahwa data ini tidak memiliki outlier.

- Jumlah Sekolah SMP (jumlah_sekolah_smp_encod): Sama seperti variabel lainnya, jumlah sekolah SMP memiliki distribusi data yang simetris tanpa outlier yang menonjol, menandakan bahwa nilai-nilai variabel ini tidak ada yang sangat menyimpang.

- Usia 15-64 (usia_15_64_encod): Distribusi data untuk kelompok usia 15-64 tahun cukup merata tanpa adanya outlier. Rentang data yang lebar menunjukkan variasi jumlah populasi di kelompok usia ini antar wilayah.

- Indihome (indihome_encod): Pada variabel ini, terdapat outlier di bagian bawah (nilai 0) sementara sebagian besar data berkumpul di nilai tinggi (sekitar 1). Hal ini mungkin menunjukkan bahwa sebagian besar wilayah memiliki akses Indihome, namun ada beberapa wilayah yang tidak memiliki akses tersebut.

- Lokal (lokal_encod): Distribusi data pada variabel ini sangat konsisten pada nilai sekitar 0 tanpa adanya variasi, menandakan bahwa data ini bersifat konstan atau biner (1 atau 0). Hal ini menunjukkan bahwa hampir semua wilayah memiliki nilai yang sama untuk variabel ini, dengan sangat sedikit atau tanpa variasi.

```python
import numpy as np

def find_outliers_iqr(data):
  Q1 = np.percentile(data, 25)
  Q3 = np.percentile(data, 75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  outliers = [x for x in data if x < lower_bound or x > upper_bound]
  return outliers

for column in dataolahencoding.columns:
  outliers = find_outliers_iqr(dataolahencoding[column])
  if outliers:
    print(f"Kolom '{column}' memiliki outlier:")
    print(outliers)
    print()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'dataolahencoding'

# 1. Standardize the data
x = dataolahencoding.values
x = StandardScaler().fit_transform(x)

# 2. Perform PCA
pca = PCA(n_components=2)  # Reduce to 2 principal components
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])

# 3. Create the Biplot
def biplot(score, coeff, labels=None):
    xs = score[:,0]
    ys = score[:,1]
    n = coeff.shape[0]
    scalex = 1.0/(xs.max() - xs.min())
    scaley = 1.0/(ys.max() - ys.min())
    plt.scatter(xs * scalex, ys * scaley)
    for i in range(n):
        plt.arrow(0, 0, coeff[i,0], coeff[i,1],color = 'r',alpha = 0.5)
        if labels is None:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, "Var"+str(i+1), color = 'g', ha = 'center', va = 'center')
        else:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, labels[i], color = 'g', ha = 'center', va = 'center')
    plt.xlabel("PC{}".format(1))
    plt.ylabel("PC{}".format(2))
    plt.grid()

biplot(principalComponents[:,0:2],np.transpose(pca.components_[0:2, :]),list(dataolahencoding.columns))
plt.show()
```

Plot PCA ini menunjukkan distribusi data berdasarkan dua komponen utama (PC1 dan PC2) serta kontribusi setiap variabel terhadap komponen tersebut. Panah merah menunjukkan arah dan kekuatan kontribusi variabel; semakin panjang panah, semakin besar pengaruh variabel tersebut dalam ruang komponen utama. Variabel-variabel seperti "jumlah sekolah smp sma encod" saling berdekatan dan searah, menunjukkan korelasi positif antarvariabel tersebut, sedangkan variabel "luas wilayah encod" memiliki arah yang berbeda, menunjukkan bahwa ia merepresentasikan aspek yang berbeda dari dataset. Secara keseluruhan, plot ini membantu mengidentifikasi variabel-variabel yang paling berpengaruh serta pola korelasi.

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
```

```python
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(dataolahencoding)

input_dim = scaled_data.shape[1]
encoding_dim = 2

input_layer = Input(shape=(input_dim,))
d_encoded = Dense(encoding_dim, activation='relu')(input_layer)
d_decoded = Dense(input_dim, activation='sigmoid')(d_encoded)

autoencoder = Model(input_layer, d_decoded)
encoder = Model(input_layer, d_encoded)

autoencoder.compile(optimizer='adam', loss='mse')

epoch = 150
autoencoder.fit(scaled_data, scaled_data, epochs=epoch, batch_size=32, shuffle=True)

encoded_data = encoder.predict(scaled_data)

encoded_df = pd.DataFrame(encoded_data, columns=['encoded_feature_1', 'encoded_feature_2'])

print(encoded_df.head())
```
